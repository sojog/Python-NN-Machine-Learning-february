```python
new_df = pd.concat([big_teams_df, df])
new_df:pd.DataFrame = new_df.reset_index(drop=True)
new_df
```

```python
new_df = new_df.sort_values(by='trophies', ascending=False).reset_index(drop=True)
new_df.head(4)
```
