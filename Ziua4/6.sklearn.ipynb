```python
new_x, new_y =  make_regression(n_samples=20, n_features=1, n_targets=1)
plt.scatter(new_x, new_y)
plt.show()
```

```python
new_x, new_y =  make_regression(n_samples=20, n_features=1, n_targets=1, noise=10)
plt.scatter(new_x, new_y)
plt.show()
```

```python
new_model = LinearRegression()
new_model.fit(new_x, new_y)
```

```python
predicted_y = new_model.predict(new_x)
```

```python
plt.scatter(new_x, new_y)
plt.plot(new_x, predicted_y)
plt.show()
```

```python
real_vs_prezis = {"y_real":new_y, "y_prezis":predicted_y}
df_real_vs_prezis = pd.DataFrame(real_vs_prezis)
df_real_vs_prezis
```
