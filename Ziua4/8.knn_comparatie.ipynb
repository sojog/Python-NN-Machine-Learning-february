```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
x_total, y_total =  make_regression(n_samples=20, n_features=1, n_targets=1, noise=10)
plt.scatter(x_total, y_total)
plt.show()
```

```python
# 80% / 20%  -> 16 4 

x_train = x_total[:16] # primele 16
y_train = y_total[:16] # primele 16

x_test = x_total[16:] # ultimele 4 (daca sunt 20 in total)
y_test = y_total[16:] # ultimele 4 (daca sunt 20 in total)
```

```python
KNN_options = range(2, 11)

score_dict = {

}

for option in KNN_options:
    knn_model = KNeighborsRegressor(n_neighbors=option)
    knn_model.fit(x_train, y_train)
    score = knn_model.score(x_test, y_test)
    score_dict[option] = score

print(score_dict)
```
