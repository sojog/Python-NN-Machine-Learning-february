```python
x = [10, 20, 30, 40, 50]
y = [10, 20, 30, 40, 50] ## y - lista de valori reale


def functie_liniara(x):
    # a  = 1 , b = 0
    return x

model = list(map(functie_liniara, x)) # model - lista de valori estimate (gasite de catre modelul meu)
plt.scatter(x, y)
plt.xlabel("Coordonata x - metri patrati")
plt.ylabel("Coordonata y - pret apartament")
plt.plot(x, model)
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [10, 20, 30, 40, 50] ## y - lista de valori reale


def functie_liniara(x):
    # a  = 1 , b = 0
    return x

model = list(map(functie_liniara, x)) # model - lista de valori estimate (gasite de catre modelul meu)
plt.scatter(x, y, c="red")
plt.xlabel("Coordonata x - metri patrati")
plt.ylabel("Coordonata y - pret apartament")
plt.plot(x, model)
plt.show()
```

```python
x = [10, 20, 30, 40, 50]
y = [10, 20, 30, 40, 50] ## y - lista de valori reale


def functie_liniara(x):
    # a  = 1 , b = 0
    return x + 1

model = list(map(functie_liniara, x)) # model - lista de valori estimate (gasite de catre modelul meu)
plt.scatter(x, y, c="red")
plt.xlabel("Coordonata x - metri patrati")
plt.ylabel("Coordonata y - pret apartament")
plt.plot(x, model)
plt.show()
```
