```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
boston_df = pd.read_csv("Boston.csv", index_col=0)
boston_df
```

```python
y_total = boston_df["medv"]
print(type(y_total), len(y_total))
```

```python
x_total = boston_df.drop(axis=1, columns=["medv"])
x_total
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_total, y_total, train_size=0.8)
for i in x_train, x_test, y_train, y_test:
    print(len(i))
```

```python
model = KNeighborsRegressor(n_neighbors=4)
model
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn_range =  range(2, 11)


knn_dict = { 
    "n_neighbors": knn_range
}
```

```python
model = KNeighborsRegressor()
```

```python
grid_model = GridSearchCV(model, knn_dict) 
grid_model
```

```python
grid_model.fit(x_train, y_train)
```

```python
knn_result_model = grid_model.best_estimator_
knn_result_model
```
